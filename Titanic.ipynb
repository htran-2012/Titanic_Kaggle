```python
df = df.drop(['PassengerId','Cabin', 'Ticket'], axis = 1)
df.shape
df.head()
```

```python
df['Family_size'] = df['SibSp'] + df['Parch'] + 1
label = 'Survived'
cat_var = ['Pclass', 'Sex', 'Embarked', 'SibSp', 'Parch', 'Family_size']
num_var = ['Age', 'Fare']
```

```python
#Count value in each categorical var
for var in cat_var:
    print(' Value counts for variable ' + var)
    print(df[var].value_counts(sort = False))
    print('-----')
```

```python
#Plot number of value in each categorical var
def plot_cate_var(var , data):
    g = sns.catplot(x = var, data =  data, kind = 'count')
    g.fig.suptitle('Count number of ' + var)
    plt.show()
    
    
for var in cat_var:
    plot_cate_var(var,df)
```

```python
# Describe numerical varue
df[num_var].describe().T
```

```python
#Plotting numerical variable distribution
def plot_dist_num_var(var, data,bins = 50, hist = True):
    sns.distplot(data[var])
    plt.show()
    
for var in num_var:
    plot_dist_num_var(var, df)
```
